# MODEL IMPLEMENTATION:

```python
import os
import torch
import torch.nn as nn
import numpy as np

from torchvision.models import resnet18, ResNet18_Weights
```

```python
from torchvision.models import resnet18
from torch.nn.modules.conv import Conv2d

class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        pretrained_model = resnet18(pretrained=True)
        self.backbone = nn.Sequential(*list(pretrained_model.children())[:-2])

        self.have_object = nn.Sequential(nn.Conv2d(512, 1, kernel_size=1), nn.AdaptiveAvgPool2d(1), nn.Sigmoid())
        self.cat_or_dog = nn.Sequential(nn.Conv2d(512, 2, kernel_size=1), nn.AdaptiveMaxPool2d(1), nn.Softmax(dim=1))
        self.specie = nn.Sequential(nn.Conv2d(512, 9, kernel_size=1), nn.AdaptiveAvgPool2d(1), nn.Softmax(dim=1))
        self.bbox = nn.Sequential(nn.Conv2d(512, 4, kernel_size=1), nn.AdaptiveAvgPool2d(1), nn.Sigmoid())

    def forward(self, input):
        out_backbone = self.backbone(input)

        bbox = self.bbox(out_backbone).view(-1, 4) * torch.tensor([1, 1, 2, 2], dtype=torch.float)
        object_presence = self.have_object(out_backbone).view(-1, 1)
        cat_or_dog = self.cat_or_dog(out_backbone).view(-1, 2)
        specie = self.specie(out_backbone).view(-1, 9)

        return {
            "bbox": bbox,
            "object": object_presence,
            "cat_or_dog": cat_or_dog,
            "specie": specie
        }

```

# CUSTOM DATALOADER IMPLEMENTATION

```python
train_list = np.load('/content/assessment_dataset/train_list.npy', allow_pickle=True).tolist()
val_list = np.load('/content/assessment_dataset/val_list.npy', allow_pickle=True).tolist()
def iou_metric(pred_boxes, gt_boxes):
    # Calculate the intersection area
    inter_xmin = torch.max(pred_boxes[:, 0], gt_boxes[:, 0])
    inter_ymin = torch.max(pred_boxes[:, 1], gt_boxes[:, 1])
    inter_xmax = torch.min(pred_boxes[:, 2], gt_boxes[:, 2])
    inter_ymax = torch.min(pred_boxes[:, 3], gt_boxes[:, 3])

    inter_width = torch.clamp(inter_xmax - inter_xmin, min=0)
    inter_height = torch.clamp(inter_ymax - inter_ymin, min=0)
    inter_area = inter_width * inter_height

    # Calculate the union area
    pred_area = (pred_boxes[:, 2] - pred_boxes[:, 0]) * (pred_boxes[:, 3] - pred_boxes[:, 1])
    gt_area = (gt_boxes[:, 2] - gt_boxes[:, 0]) * (gt_boxes[:, 3] - gt_boxes[:, 1])
    union_area = pred_area + gt_area - inter_area

    # Calculate IoU
    iou = inter_area / union_area
    return iou
```

```python
from bs4 import BeautifulSoup

def read_xml_file(path):
    try:
        with open(path, 'r') as f:
            data = f.read()
        bs_data = BeautifulSoup(data, 'xml')
        return {
            "cat_or_dog": bs_data.find("name").text,
            "xmin": int(bs_data.find("xmin").text),
            "ymin": int(bs_data.find("ymin").text),
            "xmax": int(bs_data.find("xmax").text),
            "ymax": int(bs_data.find("ymax").text),
            "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
        }
    except FileNotFoundError:
        return None
```

```python
from PIL import Image
import os
import torchvision.transforms as transforms

class CustomDataset():

    def __init__(self, root, dataset_path, images_list, train=False):

        self.root = root
        self.dataset_path = dataset_path
        self.train = train
        self.images_data = []

        image_folder_path = os.path.join(dataset_path, "images")
        label_folder_path = os.path.join(dataset_path, "labels")

        for path in os.listdir(image_folder_path):
            name = path.split(os.sep)[-1].split(".")[0]
            if name in images_list:
                xml_path = os.path.join(label_folder_path, name+".xml")
                if os.path.isfile(xml_path):
                    self.images_data.append((path, xml_path))

        if train:
            self.transform = transforms.Compose([
                transforms.Resize((224, 224)),
                transforms.RandomHorizontalFlip(),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])
        else:
            self.transform = transforms.Compose([
                transforms.Resize((224, 224)),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])

    def __getitem__(self, index):
      # Load Image
      img_path = self.images_data[index][0]
      img_path = os.path.join(self.dataset_path, "images", img_path)
      img = Image.open(img_path).convert('RGB')

      # Load Labels
      label_path = self.images_data[index][1]
      has_object = 0  # Initialize to 0 in case an exception is thrown
      xmin, ymin, xmax, ymax = 0, 0, 0, 0  # Initialize to 0 in case an exception is thrown
      try:
          with open(label_path, "r") as f:
              label = f.readline()
              label_parts = label.split()
              has_object = int(label_parts[0])
              if has_object:
                  specie = int(label_parts[1])  # Assuming the 'specie' label is present in the label file
                  cat_or_dog = int(label_parts[2])
                  xmin, ymin, xmax, ymax = map(int, label_parts[3:])
              else:
                  specie = 0
                  cat_or_dog = 0
      except ValueError:
          specie = 0
          cat_or_dog = 0

      # Convert the labels to tensors and change their data types to long
      has_object = torch.tensor(has_object, dtype=torch.float)
      specie = torch.tensor(specie, dtype=torch.float)
      cat_or_dog = torch.tensor(cat_or_dog, dtype=torch.float)
      xmin = torch.tensor(xmin, dtype=torch.float)
      ymin = torch.tensor(ymin, dtype=torch.float)
      xmax = torch.tensor(xmax, dtype=torch.float)
      ymax = torch.tensor(ymax, dtype=torch.float)
      # Return Tuple (img, label)
      return self.transform(img), {"has_object": has_object, "specie": specie, "cat_or_dog": cat_or_dog,"xmin":xmin, "ymin":ymin,"xmax":xmax,"ymax":ymax}



    def __len__(self):
        return len(self.images_data)
```
